```python
import pandas as pd
import numpy as np

df = pd.read_excel("https://raw.githubusercontent.com/ywchiu/tibame_tm/master/data/google_comments.xlsx", index_col=0)
df.head()
```

```python
def convert_score_to_status(score):
    if int(score) >= 4:
        return 'good'
    elif int(score) <= 2:
        return 'bad'
    else:
        return 'normal'
```

```python
df['status'] = df['score'].map(lambda e: convert_score_to_status(e))
df.head()
```

```python
df = df[['comment', 'status']]
```

```python
df.head()
```

```python
def convert_status_to_label(s):
    if s == 'good':
        return 1
    else:
        return 0
```

```python
df['label'] = df['status'].map(lambda e: convert_status_to_label(e))
df.head()
```

```python
df = df[['comment', 'label']]
df.head()
```

```python
from sklearn.model_selection import train_test_split
train_df, eval_df = train_test_split(df, test_size=0.2, random_state=42)
```

```python
eval_df.shape
```

```python
from simpletransformers.classification import ClassificationModel
model = ClassificationModel('bert', 'bert-base-chinese', use_cuda=False)

```

```python
from opencc import OpenCC
cc = OpenCC('t2s.json')
```

```python
train_df['comment'] = train_df['comment'].map(lambda e: cc.convert(e))
eval_df['comment'] = eval_df['comment'].map(lambda e: cc.convert(e))
```

```python
model.eval_model(eval_df)
```

```python
model.train_model(train_df, args = {'overwrite_output_dir':True})
```

```python
result, model_outputs, wrong_preds = model.eval_model(eval_df)
```
