## Crawler

```python
import requests
from bs4 import BeautifulSoup

headers = {
    'user-agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"
}
url = "https://www.ptt.cc/bbs/creditcard/index.html"
res = requests.get(url, headers=headers)
soup = BeautifulSoup(res.text, 'lxml')
# print(res.text)

corpus = []
for item in soup.select('div .title'):
  corpus.append("".join(item.text.strip().split()))
print(corpus)
```

##Jieba

```python
import jieba
jieba_corpus = []
for sentence in corpus:
  jieba_corpus.append(" ".join([w for w in jieba.lcut(sentence) if len(w) >= 2]))
print(jieba_corpus)
```

## Count Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
vec = CountVectorizer()
vec_ary = vec.fit_transform(jieba_corpus).toarray()
# print(vec.get_feature_names_out())
# print(vec_ary)

term_ary = vec.get_feature_names_out()
for c_matrix in vec_ary:
  print([(i,j) for i,j in zip(term_ary, c_matrix)])
```

##TF-IDF Vectorizer

```python
vec = TfidfVectorizer()
vec_ary = vec.fit_transform(jieba_corpus).toarray().round(3)
# print(vec.get_feature_names_out())
# print(vec_ary)

term_ary = vec.get_feature_names_out()
for t_matrix in vec_ary:
  print([(i,j) for i,j in zip(term_ary, t_matrix)])
```
