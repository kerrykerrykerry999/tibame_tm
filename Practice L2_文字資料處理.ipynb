## Generator (補充資料)

generator是一種特別的iterator，可以透過yield 讓我們自行控制iterator 的狀態，決定是否要呼叫__next__讓他進行迭代行為。Generator 採取Lazy Evaluation，真正需要資料時才進行計算。

優點：

1. 不需要定義__next，提高程式碼可讀性
2. 減少記憶體的使用
3. 可以用'yield from'實作遞迴的迭代行為
4. 可以建立一個可拆式的資料管線，減少了不同工作階段的耦合性

```python
import jieba
from collections import Counter
import requests
from bs4 import BeautifulSoup

jieba.load_userdict("sample_dict.txt")

headers = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"}
res = requests.get("https://autos.udn.com/autos/story/8324/6710738?from=udn-indexlistnews_ch1018", headers=headers)
# print(res.text)

paragraph_word_list = []
paragraph = ""
soup = BeautifulSoup(res.text, 'lxml')
for item in soup.select("p"):
  paragraph = paragraph + item.text

# 不需要放到list，jieba.cut()的return本來就是可迭迨object
# paragraph_word_list = list(jieba.cut(paragraph))
# print(paragraph_word_list)

## Non-Pythonic
remain_word_list=[]
for word in jieba.cut(paragraph):
  if len(word)>=2:
    remain_word_list.append(word)
c = Counter(remain_word_list)
print(c)

## Pythonic
c = Counter([w for w in jieba.cut(paragraph) if len(w)>=2])
print(c)

```
